```python
import seaborn as sns
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["time"]=le.fit_transform(df["time"])
```

```python
df["time"]
```

```python
df.head()
```

```python
x=df.drop(labels="time",axis=1)
y=df["time"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
numcol=["total_bill","tip","size"]
catcol=["sex","smoker","day"]
```

```python
num_pipeline = Pipeline([
    # Step 1: Impute missing values
    ('imputer', SimpleImputer(strategy='median')),  # You can choose median, mean, or any other strategy

    # Step 2: Scale features
    ('scaler', StandardScaler()),  # Standardize features by removing the mean and scaling to unit variance
])

cat_pipeline = Pipeline([
    # Step 1: Impute missing values
    ('imputer', SimpleImputer(strategy="most_frequent", fill_value='missing')),  # Fill missing values with a constant value

    # Step 2: One-hot encoding
    ('encoder', OneHotEncoder()),  # Drop first column to avoid multicollinearity (dummy variable trap)
])

# Fit and transform categorical columns
preprocessor = ColumnTransformer([
    ('num', num_pipeline, numcol),
    ('cat', cat_pipeline, catcol),
])
x_train_transformed = preprocessor.fit_transform(x_train)
x_test_transformed = preprocessor.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier()),  # Example model
])
pipeline.fit(x_train, y_train)
y_pred = pipeline.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
def evaluate_models(X_train, y_train, X_test, y_test, models):
    model_accuracies = {}
    for model_name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)

        # Predictions
        y_pred = model.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)

        # Save accuracy
        model_accuracies[model_name] = accuracy

    return model_accuracies
models = {
    'Random Forest Classifier': Pipeline([
        ('preprocessor', preprocessor),
        ('model', RandomForestClassifier(random_state=42)),
    ]),
    'DecisionTreeClassifier': Pipeline([
        ('preprocessor', preprocessor),
        ('model', DecisionTreeClassifier(random_state=42)),
    ]),
    'LogisticRegression': Pipeline([
        ('preprocessor', preprocessor),
        ('model', LogisticRegression()),  # Replace NewModel() with your new model instance
    ])
}

# Evaluate models
model_accuracies = evaluate_models(x_train, y_train, x_test, y_test, models)

# Print model accuracies
for model_name, accuracy in model_accuracies.items():
    print(f"{model_name} Accuracy: {accuracy}")
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Define columns to be one-hot encoded
columns_to_encode = ['sex', 'smoker', 'day']

# Apply one-hot encoding to categorical columns
encoder = OneHotEncoder(drop='first', sparse=False)
x_train_encoded = encoder.fit_transform(x_train[columns_to_encode])

# Get the feature names for the encoded columns
encoded_columns_names = encoder.get_feature_names_out(columns_to_encode)

# Convert the encoded array into a DataFrame
x_train_encoded_df = pd.DataFrame(x_train_encoded, columns=encoded_columns_names)

# Drop the original categorical columns from x_train and concatenate the encoded columns
x_train_processed = pd.concat([x_train.drop(columns=columns_to_encode), x_train_encoded_df], axis=1)

# Now, x_train_processed contains the one-hot encoded features
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Define the Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=42)

# Define the hyperparameters grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(x_train_encoded_df,y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Negative Mean Squared Error:", best_score)
```
